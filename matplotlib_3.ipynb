### Plot Appearance

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,11)
y = x**2
```

* alpha is use for the transparency of the line
* linewidth will also work on the lw in the ax.plot()
means that 
"ax.plot(x,y,linewidth=3)"
"ax.plot(x,y,lw=3)"
both are same 
* linestyle has multiple attributes such as --, :, steps,
* marker also has a multiple output such as o, O, ^, v, >, <, +, x, X etc.


```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'red',linewidth=2,alpha=0.7,linestyle='--',marker='o',markersize=5, markerfacecolor = 'yellow',markeredgewidth=1,markeredgecolor='blue')
#ax.plot(x,y,color='purple',lw=3,ls=':')
```

### Plot Range

* ax.set_xlim([0,1]) gives the range of plot from a certain range of given data input, here x will plot by 2 index only
* ax_setylim([0,2]) works similarly in y axis

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'purple',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
