```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
# g.map(plt.scatter) this is done for the total plotting in pairplot
g.map_diag(sns.histplot) #diagonal of pairplot
# here distplot will not work, instead using the histplot
g.map_upper(sns.scatterplot) #upper half of pairplot from diagonal
g.map_lower(sns.kdeplot) #lower half of pairplot from diagonal
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
g = sns.FacetGrid(data=tips,row='time',col='smoker')
g.map(sns.histplot,'total_bill')
g.map(plt.scatter,'total_bill','tip')
```
