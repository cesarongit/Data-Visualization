### Plotly and Cufflinks

```python
#cufflinks connects the plotly with pandas
import pandas as pd
import numpy as np
```

```python
from plotly import __version__
```

```python
print (__version__)
#pierian was using the 1.12.9
```

```python
import cufflinks as cf 
from plotly.offline import download_plotlyjs,init_notebook_mode,iplot
```

```python
init_notebook_mode(connected=True) #this connect the javascript to notebook
```

```python
cf.go_offline()
```

```python
#data
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2
```

```python
df.iplot() #iplot is used instead of plot
#this connects the pandas to the plotly, which results in direct plotly making a line plot for us without calling actual plotly
```

```python
df.iplot(kind='scatter',x='A',y='B',mode='markers',size=10)
```

```python
df2.iplot(kind='bar',x='Category',y='Values')
```

```python
df.sum().iplot(kind='bar') #sum of the integers present in the columns
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3.iplot(kind='surface',colorscale='rdylbu')
```

```python
df3.iplot(kind='scatter',mode='markers')
```

```python
df.iplot(kind='hist',bins=50)
```

```python
df[['A','B']].iplot(kind='spread') 
```

```python
df.iplot(kind='bubble',x='A',y='B',size='C',color='orange')
```

```python
df.scatter_matrix()
```
