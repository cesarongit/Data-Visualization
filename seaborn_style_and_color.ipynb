```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('ticks')
sns.countplot(x ='sex',data=tips)
sns.despine(bottom=True,left=True) #right and top are by default true
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x ='sex',data=tips)
```

```python
sns.set_context('talk') #default is notebook
sns.countplot(x ='sex',data=tips)
```

```python
g = sns.FacetGrid(data=tips,row='total_bill',col='tip')
g.map(sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic',x_ci='sd'))

g.map(sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic',x_ci='sd'))
```
