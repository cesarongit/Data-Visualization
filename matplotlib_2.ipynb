```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
```

```python
x = np.linspace(0,5,11)
y = x**2
fig,axes = plt.subplots(nrows=1,ncols=2)
#axes.plot(x,y)
plt.tight_layout() #- it is used to avoid the collision between the plots

for current_ax in axes:
    current_ax.plot(x,y) #we can plot this because the axes is here an array
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y) #second method to plot
axes[1].plot(y,x)
axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')
plt.tight_layout() #advices to put this in every plot to avoid the overlapping of the plot
```
